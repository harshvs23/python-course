```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('watches.csv')
df
```

```python
df.info()
```

```python
def clean_price_column(price):
    if isinstance(price, str):
        price = price.replace(',', '')
        return float(price)
    return price
```

```python
clean_price_column('239,002')
```

```python
df['price'].apply(clean_price_column) # column selection
```

```python
df['clean_price'] = df['price'].apply(clean_price_column)
```

```python
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating, str):
        if len(rating) > 3:
            return np.nan
        if rating.count('.') == 1:
            return float(rating)
        if rating.isnumeric():
            return int(rating)
        return np.nan
    return rating
```

```python
clean_rating_column('1 stock left only')
```

```python
df['rating'].apply(clean_rating_column)
```

```python
df['clean_rating'] = df['rating'].apply(clean_rating_column)
```
